```python
import pandas as pd
pd.set_option("max_colwidth", None)
pd.set_option("max_seq_items", None)
```

```python
coach_df = pd.read_csv("./csv/coach_data.csv", sep="\t", encoding="utf-16", parse_dates=["Date","Match Time","Month"])
coach_df.sort_values(by='Match Time', ascending = False, inplace = True)
coach_df.rename(columns={"T Ds": "TDs"}, inplace = True)
```

```python
# cols_to_drop = ["Date", "Description", "Game Description", "Month", "Tournament Name", "Month Filter - Coach Data", 
#                 "Year Filter - Coach Data", "Number of Records", "Most Recent", "URL", "Number of Games (Coach)", 
#                 "Number of Games (Tournament)", "Number of Games (Tournament) (copy)", "Rank (copy)", 'Race In', 
#                 'Race Initial', 'Race Label', 'Race Colour']

cols_to_keep = ['Coach', 'Opponent', 'Nation', 'Opponent Nation', 
       'Location', 'Match Id', 'Match Time', 'Variant',
       'Race', 'Opponent Race',
                 'Rank', 'Opponent Rank', 
                'Result',  'Opponent Result',
                'TDs','Opponent TDs', 
       'Seat', 'Tier',
       'Win %']

coach_df = coach_df[cols_to_keep]

print(coach_df.columns)
coach_df.head(5)
```

```python
coach_df.to_csv("./csv/coach_data_cleaned.csv", sep="\t", encoding="utf-8", index=False)
```

```python
# get data ready for ML
coach_df = coach_df[coach_df["Variant"] == 'BB 2020']

coach_df['Result'] = coach_df['Result'] == "W"
coach_df['Result'] = coach_df['Result'].replace({True: 1, False: 0})
coach_df['Seat'] = coach_df['Seat'] == "Home"
coach_df['Seat'] = coach_df['Seat'].replace({True: 1, False: 0})
coach_df['Rank Difference'] = coach_df['Opponent Rank'] - coach_df['Rank']

cols_to_keep = ['Race', 'Opponent Race',
                 'Rank Difference', 
              'Seat','Result']


coach_df = coach_df[cols_to_keep]
coach_df
```

```python
coach_df.to_csv("./csv/coach_data_prepared_for_ml.csv", sep="\t", encoding="utf-8", index=False)
```
